# Introduction

```python
# switch keras to use tensorflow 
import os
os.environ["KERAS_BACKEND"] = "tensorflow"
```

```python
# import keras
from __future__ import print_function
import keras
import tensorflow as tf
from keras.models import Sequential
from keras.layers import Dense, Dropout, Flatten
from keras.layers import Conv2D, MaxPooling2D

# import visualization tools
import matplotlib as mpl
from keras.utils import plot_model
from matplotlib.pyplot import imshow
import matplotlib.pyplot as plt
import numpy as np
```

# Dataset

Keras has a build-in [daset](https://keras.io/datasets) such as [CIFAR10](https://www.cs.toronto.edu/~kriz/cifar.html), [MNIST](http://yann.lecun.com/exdb/mnist/) and [Boston housing](https://www.kaggle.com/vikrishnan/boston-house-prices). This make it very easy to try out keras

# MNIST
 [MNIST](http://yann.lecun.com/exdb/mnist/) is a image's hand writing data set. 
  
The MNIST database of handwritten digits, available from this page, has a training set of 60,000 examples, and a test set of 10,000 examples. It is a subset of a larger set available from NIST. The digits have been size-normalized and centered in a fixed-size image.
It is a good database for people who want to try learning techniques and pattern recognition methods on real-world data while spending minimal efforts on preprocessing and formatting.

```python
# import dataset
from keras.datasets import mnist
# the data, split between train and test sets
(x_train, y_train), (x_test, y_test) = mnist.load_data()
```

```python
# view current shape
print("x_train shape: ", x_train.shape)
print("y_train shape: ", y_train.shape)
print("x_test shape: ", x_test.shape)
print("y_test shape: ", y_test.shape)
```

```python
# show first 20 images
fig=plt.figure(figsize=(20, 2))
columns = 20
rows = 2
for i in range(0, columns*rows):
    fig.add_subplot(rows, columns, i+1)
    plt.imshow(x_train[i], cmap='gray')
print("=====x_train=====")
plt.show()
print("=====y_train=====")
print(y_train[:20])
print(y_train[20:40])
```

```python
# view sample
print(x_train[0])
```

# Hyper Parameter

```python
# config
batch_size = 128 # number of training data per time (จำนวนรุปที่รับเข้าไป train ใน แต่ละรอบ)
epochs = 12 # number of trainning cycle (จำนวนรอบในการ train)

```

## Preprocess data
2 goals:
- The x data is 2d array of number between 0-255. It is best to normalize them to 0-1.
- The input should be in width, height, color chanel

```python
# input image dimensions
img_rows, img_cols = 28, 28 # image height and width
x_train = x_train.reshape(x_train.shape[0], img_rows, img_cols, 1)
x_test = x_test.reshape(x_test.shape[0], img_rows, img_cols, 1)
# view current shape
print("x_train shape: ",x_train.shape)
print("y_train shape: ",y_train.shape)
print("x_test shape: ",x_test.shape)
print("y_test shape: ",y_test.shape) 
```

```python
print(x_train[0])
# visualize each element
for x_train_row in x_train[0]:
    for x_train_column in x_train_row:
        print(x_train_column,end=' ')
    print()
```

![](images/reshape.png)

```python
# preprocess
x_train = x_train.astype('float32')
x_test = x_test.astype('float32')
x_train /= 255
x_test /= 255

# visualize
print('x_train shape:', x_train.shape)
print(x_train[0])
# visualize each element
for x_train_row in x_train[0]:
    for x_train_column in x_train_row:
        print(x_train_column,end=' ')
    print()
```

```python
# convert class vectors to binary class matrices
num_classes = 10 # number of output in this case: 10 digit (จำนวนข้อมูลออก: 10 เลข)
y_train = keras.utils.to_categorical(y_train, num_classes)
y_test = keras.utils.to_categorical(y_test, num_classes)
# visualize
print(y_train.shape)
print(y_train[0])
```

![](images/categorical.png)

## After preprocessing
![](images/input_output.png)

After preprocess we will get a x and y that is compatible with the model.

## create a model

```python
input_shape = (img_rows, img_cols, 1)
model = Sequential()
```

![](images/sequential.png)

```python
model.add(Conv2D(32, kernel_size=(3, 3),
                 activation='relu',
                 input_shape=input_shape))
model.add(Conv2D(64, kernel_size==(3, 3), activation='relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Dropout(0.25))
model.add(Flatten())
model.add(Dense(128, activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(num_classes, activation='softmax'))
print(model.summary())
```

## Compile the model

```python
model.compile(loss=keras.losses.categorical_crossentropy,
              optimizer=keras.optimizers.Adadelta(),
              metrics=['accuracy'])
```

# Train the model

```python
model.fit(x_train, y_train,
          batch_size=batch_size,
          epochs=epochs,
          verbose=1,
          validation_data=(x_test, y_test))
```

# Evaluate the model

```python
score = model.evaluate(x_test, y_test, verbose=0)
print('Test loss:', score[0])
print('Test accuracy:', score[1])
```
